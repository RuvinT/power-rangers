```python
# basic libraries
import pandas as pd
```

```python
# getting datasets from 2002 to 2023
datasets = []
for i in range(2002, 2024):
    path = "D:\Study\AISC 2006 - Step Presentation 02\Datasets\IESO Historical Demand for Electricity\PUB_Demand_{}.csv".format(i)
    datasets.append(pd.read_csv(path, skiprows=3))
```

```python
# check datasets
for dataset in datasets:
    print(dataset)
```

```python
# row-wise concatenate all datasets
final_dataset = pd.concat(datasets, axis=0)
```

```python
# check final combined dataset
final_dataset
```

```python
# create final dataset csv file
final_dataset.to_csv("D:\Study\AISC 2006 - Step Presentation 02\Datasets\IESO Historical Demand for Electricity\PUB_Demand.csv", index=False)
```
