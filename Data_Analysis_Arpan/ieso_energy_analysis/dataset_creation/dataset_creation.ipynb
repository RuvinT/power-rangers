```python
# basic libraries
import pandas as pd
```

1. IESO Demand Datasets

```python
# getting datasets from 2002 to 2023
demand_datasets = []
for i in range(2002, 2024):
    path = "Datasets\IESO Historical Demand for Electricity\PUB_Demand_{}.csv".format(i)
    demand_datasets.append(pd.read_csv(path, skiprows=3))
```

```python
# check datasets
for dataset in demand_datasets:
    print(dataset)
```

```python
# row-wise concatenate all datasets
final_demand_dataset = pd.concat(demand_datasets, axis=0)
```

```python
# check final combined dataset
final_demand_dataset
```

```python
# create final dataset csv file
final_demand_dataset.to_csv("Datasets\Final Datasets\ieso_demand_dataset.csv", index=False)
```

2. IESO Supply Datasets

```python
# getting datasets from 2015 to 2023
supply_datasets = []
for i in range(2015, 2024):
    path = "Datasets\IESO Generator Output by Fuel Type Hourly Report\Generator_Supply_{}.csv".format(i)
    supply_datasets.append(pd.read_csv(path))
```

```python
# check datasets
for dataset in supply_datasets:
    print(dataset)
```

```python
# row-wise concatenate all datasets
final_supply_dataset = pd.concat(supply_datasets, axis=0)
```

```python
# check final combined dataset
final_supply_dataset
```

```python
# create final dataset csv file
final_supply_dataset.to_csv("Datasets\Final Datasets\ieso_supply_dataset.csv", index=False)
```

3. IESO Price Datasets

```python
# getting datasets from 2002 to 2023
price_datasets = []
for i in range(2002, 2024):
    path = "Datasets\IESO Yearly Hourly HOEP or Predispatch Report\PUB_PriceHOEPPredispOR_{}.csv".format(i)
    price_datasets.append(pd.read_csv(path, skiprows=3))
```

```python
# check datasets
for dataset in price_datasets:
    print(dataset)
```

```python
# row-wise concatenate all datasets
final_price_dataset = pd.concat(price_datasets, axis=0)
```

```python
# check final combined dataset
final_price_dataset
```

```python
# create final dataset csv file
final_price_dataset.to_csv("Datasets\Final Datasets\ieso_price_dataset.csv", index=False)
```

Combination of All 3 Datasets

```python
# merge demand and supply datasets based on date and hour
final_demand_supply_dataset = pd.merge(final_demand_dataset, final_supply_dataset, how='outer', left_on=['Date', 'Hour'], right_on=['Date', 'Hour'])
```

```python
# merge demand and supply and price datasets based on date and hour
final_demand_supply_price_dataset = pd.merge(final_demand_supply_dataset, final_price_dataset, how='outer', left_on=['Date', 'Hour'], right_on=['Date', 'Hour'])
final_dataset = final_demand_supply_price_dataset
```

```python
# check final dataset
final_dataset
```

```python
# create final dataset csv file
final_dataset.to_csv("Datasets\Final Datasets\ieso_dataset.csv", index=False)
```
